##  Observations

• Females survived more than males
• First-class passengers had higher survival rates
• Passengers with higher fares had better survival chances
• Younger passengers survived slightly more
• Age and Cabin columns have missing values


## Summary

The exploratory data analysis of the Titanic dataset shows that survival was strongly influenced by gender, passenger class, and fare. Females and first-class passengers had significantly higher survival rates. Fare showed a positive correlation with survival, while age had a moderate effect. Missing values were observed mainly in the Age and Cabin columns.
